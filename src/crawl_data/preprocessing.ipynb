```python
import pandas as pd
import os
import re
```

```python
laptop_file = ["laptop.csv", "laptop1.csv", "laptop3.csv"]
wranty_file = ["csbh.txt", "csdt.txt", "csvt.txt"]
```

```python
df_laptop = []
for i in range(len(laptop_file)):
    df_laptop.append(pd.read_csv(laptop_file[i]))
    # df_laptop[i] = df_laptop[i][:39].copy()

for df in df_laptop:
    df.value_counts()
```

```python
df_unique = []
for df in df_laptop:
    df_unique.append((df.drop_duplicates())[:30])
    
```

```python
def cleaning_text(text : str)-> str:
    """
     loai bo cac ki tu xuong dong khi chua ket thuc cau, xoa khoang trong thua.
    """
    # thay xuong dong thanh khoang trang khi chua ket thuc cau
    text = re.sub(r'(?<![.!?;:])\n(?!\n)', ' ', text)
    # xoa khoang trong thua
    text = re.sub(r'\n{2,}', '\n\n', text)
    # text = re.sub(r'^\s*-{3,}\s*$', '', text, flags=re.MULTILINE)
    # # 1. Xóa các khối điều kiện MSO (Microsoft Word/Outlook HTML artifacts)
    # text = re.sub(r'\[if\s+gte\s+mso\s+\d+\]>.*?<!\[endif\]', '', text, flags=re.DOTALL)
    
    return text.strip()
```

```python
laptop_ouput = r"final_data/laptop.txt"
final_text =[]

for df in df_unique:
    links = (df.iloc[:, 0].astype(str))
    prices = (df.iloc[:, 1].astype(str))
    descs =(df.iloc[:,2].astype(str))

    line = [ f"giá sản phẩm đang được khuyến mãi là: {price}, chi tiết mô tả sản phẩm tham khảo tại đây: {link}, {cleaning_text(desc)}"
            for price,link, desc in zip(prices, links, descs)]
    final_text.append("\n\n\n".join(line))

os.makedirs(os.path.dirname(laptop_ouput), exist_ok=True)
with open(laptop_ouput, "w", encoding="utf-8") as f:
    f.write("\n\n\n".join(final_text))
    
```
